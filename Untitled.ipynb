```python
from obspy.io.seg2.seg2 import SEG2
import numpy as np
import matplotlib.pyplot as plt
import obspy
import os
def read_campus_data(filenames):
    if type(filenames) is not list:
        print("Argumentet til \"read_campus_data\" skal være en liste af filnavne. Listen kan godt indeholde 1 fil, fx read_campus_data([\"100.dat\"])")
        return
    if (invalid := np.where([not os.path.exists(x) for x in filenames])[0]).any():
        print(f"Argumentet til \"read_campus_data\" skal være en liste af filnavne, men det er {np.array(filenames)[invalid]} ikke. Hvis fil(erne) er i en underfolder, skal det være en del af filnavnet. Den nuværende folder er {os.getcwd()}/")
        return
    amplitudes = np.array([[x.data for x in SEG2().read_file(filename)] for filename in filenames])
    times = SEG2().read_file(filenames[0])[0].times()
    return times, amplitudes

def read_marine_data(filename):
    file = obspy.read("kat15_line02_sel_raw.sgy")
    amplitudes = np.array([x.data for x in file]).reshape(-1,48,2401)
    times = file[0].times()
    return times, amplitudes

def show_timeline(tider, afstande, amplituder, skalaer = 1):  
    """
    Funktion til at visualisere seismometer data som en funktion af tid, efter den ønskede pre-processering
    ----------
    tider : liste 
        Liste af tiderne data er blevet taget, i sekunder. 
    afstande : liste 
        Position af hver mikrofon, i meter. 
    amplituder : 2d array
        En 2d-array med amplituder der skal plottes. 
        Hver række (1. dimension) repræsenterer en mikrofon, mens hver kolonne repræsenterer et tidspunkt
    skalaer : float
        Størrelse af svingninger i plottet for hver mikrofon.
    """
 
    tider = np.array(tider)
    afstande = np.array(afstande)
    amplituder = np.array(amplituder)
    assert amplituder.ndim == 2, f"amplituder skal være en 2-dimensionel array, men du har givet et {amplituder.ndim} dimensionelt array. Har du måske givet data fra alle forsøg samlet? så kan du få den første med indexering: amplitudes[0]"
    assert tider.ndim == 1 and afstande.ndim == 1 and skalaer.ndim == 1, f"\"tider\", \"afstande\" og \"skalaer\" skal være 1-dimensionelle, men har her formerne hhv {tider.shape}, {afstande.shape} og {skalaer.shape}"
    assert len(tider) == amplituder.shape[1], f"\"amplituder\" har {amplituder.shape[1]} kolonner og derfor tidspunkter, men \"tider\" specificerer {len(tider)} tider."
    assert len(afstande) == amplituder.shape[0], f"\"amplituder\" har {amplituder.shape[0]} rækker hvilket skal være antallet af mikrofoner, men \"afstande\" specificerer afstanden til {len(afstande)} mikrofoner."
    N = len(afstande)
    corr_amp = amplituder * skalaer[:,None]
    scaling = np.diff(afstande).mean() / np.nanmax(np.abs(corr_amp)) 
    for idx in range(N):
        plt.fill_betweenx(tider, afstande[idx] + corr_amp[idx] * scaling, afstande[idx]+0*tider, linestyle="-")
        #plt.plot(afstande[idx] + (1*tider * corr_amp), tider,linewidth=0.7)
        plt.xlabel("distance [m]")
        plt.ylabel("time [s]")
    plt.gca().invert_yaxis()
```

```python
file = obspy.read("kat15_line02_sel_raw.sgy")
amplitudes = np.array([x.data for x in file]).reshape(-1,48,2401)
times = file[0].times()
file
```

```python
from scipy.signal import butter,filtfilt

def butter_highpass_filter(data, cutoff, fs, order):
    normal_cutoff = cutoff / (0.5*fs)
    # Get the filter coefficients 
    b, a = butter(order, normal_cutoff, btype='high', analog=False)
    y = filtfilt(b, a, data)
    return y
```

```python
amplitudes_stacked = np.sum(amplitudes**2,0)**0.5
#amplitudes_stacked = np.mean(amplitudes,0)
```

```python
fs = times[1] - times[0]
```

```python
z_water = 25
v_water = 1480
z_layer1 = 25
v_layer1 = 1480
z_layer2 = 150
v_layer2 = 1800
z_layer3 = 165
v_layer3 = 2000
def surface_response(dist):
    return dist / v_water

def seafloor_response(dist):
    return 2 * np.hypot(dist/2, z_water) / v_water

def layer1_response(dist):
    z_tot = z_water+z_layer1
    d_w = dist/2*z_water/z_tot
    d_1 = dist/2*z_layer1/z_tot
    return 2 * (np.hypot(d_w, z_water)/v_water + np.hypot(d_1, z_layer1) / v_layer1)

def layer2_response(dist):
    z_tot = z_water+z_layer1+z_layer2
    d_w = dist/2*z_water/z_tot
    d_1 = dist/2*z_layer1/z_tot
    d_2 = dist/2*z_layer2/z_tot
    return 2 * (np.hypot(d_w, z_water)/v_water + np.hypot(d_1, z_layer1) / v_layer1 + np.hypot(d_2, z_layer2) / v_layer2)

def layer3_response(dist):
    z_tot = z_water+z_layer1+z_layer2+z_layer3
    d_w = dist/2*z_water/z_tot
    d_1 = dist/2*z_layer1/z_tot
    d_2 = dist/2*z_layer2/z_tot
    d_3 = dist/2*z_layer3/z_tot
    return 2 * (np.hypot(d_w, z_water)/v_water + np.hypot(d_1, z_layer1) / v_layer1 + np.hypot(d_2, z_layer2) / v_layer2 + np.hypot(d_3, z_layer3) / v_layer3)

N = amplitudes.shape[1]
plt.figure(figsize=(8,6))
dists = 6.4*np.arange(48) + 20
for idx in range(48):
    corr_amp = butter_highpass_filter(amplitudes_stacked[idx], 1, 1/fs, 1)
    plt.plot(dists[idx] + 1*times * corr_amp, times,linewidth=0.7)
    plt.xlabel("distance [m]")
    plt.ylabel("time [s]")
plt.gca().invert_yaxis()
plt.ylim(0.3, 0)
plt.plot(dists, surface_response(dists), "--k")
plt.plot(dists, seafloor_response(dists), "--k")
plt.plot(dists, layer1_response(dists), "--k")
plt.plot(dists, layer2_response(dists), "--k")
plt.plot(dists, layer3_response(dists), "--k")
plt.grid()
v = 230     #m/s
t0 = -8e-3  #s
#plt.plot(v * times,times - t0)
```
